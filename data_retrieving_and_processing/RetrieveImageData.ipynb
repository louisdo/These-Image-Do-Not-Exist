```python
def search_image(text, num_images):
    if num_images % 250 == 0: num_pages = int(num_images / 250)
    else: num_pages = int(num_images / 250) + 1

    results = []
    for page in tqdm(range(1, num_pages + 1), desc = f"Searching images, query = {text}"):
        search_results = flickr.photos.search(text = text, 
                                            extras = "url_s,views",
                                            privacy_filter = 1,
                                            safe_search = 1,
                                            sort = "relevance",
                                            content_type = 1,
                                            per_page = 250,
                                            page = page)
        root = ET.fromstring(search_results.decode("UTF-8"))

        for child in root:
            for c in child:
                try:
                    dp = {"id": c.attrib["id"], "url": c.attrib["url_s"], "views": int(c.attrib["views"])}
                    results.append(dp)
                except Exception: 
                    print(c.attrib)
                    continue

    return results

def remove_duplicate(data):
    all_ids = set([])
    res = []
    for dp in data:
        photo_id = dp["id"]
        if photo_id not in all_ids:
            res.append(dp)
            all_ids.add(photo_id)
    return res
```

```python
BIG_CATEGORIES = [["castle", "palace", "fortress", "stronghold"], 
                ["landscape", "hill", "forest", "jungle"], 
                ["sea", "seascape", "coast", "ocean"], 
                ["cityscape", "city", "skyscraper", "new york"], 
                ["cat", "house cat", "cute cat", "kitten"]]
```

```python
DATA_CATEGORIES = {i: [] for i in range(len(BIG_CATEGORIES))}
for i, category in tqdm(enumerate(BIG_CATEGORIES)):
    for keyword in category:
        search_results = search_image(text = keyword, num_images = 4000)
        DATA_CATEGORIES[i].extend(search_results)
    DATA_CATEGORIES[i] = remove_duplicate(DATA_CATEGORIES[i])
```

```python
with open("drive/MyDrive/HUST_related/Project_III/data/image_info_data.json", "w") as f:
    json.dump(DATA_CATEGORIES, f)
```
